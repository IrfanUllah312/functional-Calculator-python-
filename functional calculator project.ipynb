```python
from tkinter import *
from PIL import Image,ImageTk 
window = Tk()
window.configure(bg="black")
window.geometry("564x640")
window.title("project no 1")
window.resizable(False,False)


#Functionn
ans=""

def cal(value):
    
    global ans
    ans+=value
    l1.config(text=ans,fg="blue")    

def dell():
    global ans
    if(ans!=""):
        ans = ans[0:len(ans)-1]
        l1.config(text=ans)
        
def answer():
    global ans
    
    if(ans != ""):
        try:
            ans = eval(ans)
        except:
            ans="Error"
    
    l1.config(text=ans,fg="yellow")
    ans=""






#Answer screen
l1=Label(window,text="",width=33,height=5,bg="grey",fg="white",font="arail 20 bold")
l1.pack()


"""bbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbb
bbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbb
bbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbb"""

image1 = Image.open("del.png").resize((100,80))
pic1 = ImageTk.PhotoImage(image1)
b1=Button(window,image=pic1,command= lambda button="del":dell()).place(x=0,y=200)

image2 = Image.open("Divide.png").resize((100,80))
pic2 = ImageTk.PhotoImage(image2)
b2=Button(window,image=pic2,command=lambda button="/":cal(button)).place(x=150,y=200)

image3 = Image.open("x.png").resize((100,80))
pic3 = ImageTk.PhotoImage(image3)
b3=Button(window,image=pic3,command=lambda button="*":cal(button)).place(x=300,y=200)

image4 = Image.open("-.png").resize((100,80))
pic4 = ImageTk.PhotoImage(image4)
b4=Button(window,image=pic4,command = lambda button="-":cal(button)).place(x=450,y=200)

#second row

image5 = Image.open("7.png").resize((100,80))
pic5 = ImageTk.PhotoImage(image5)
b5=Button(window,image=pic5,command=lambda button="7":cal(button)).place(x=0,y=320)

image6 = Image.open("8.png").resize((100,80))
pic6 = ImageTk.PhotoImage(image6)
b6=Button(window,image=pic6,command=lambda button="8":cal(button)).place(x=150,y=320)

image7 = Image.open("9.png").resize((100,80))
pic7 = ImageTk.PhotoImage(image7)
b7=Button(window,image=pic7,command=lambda button="9":cal(button)).place(x=300,y=320)

image8 = Image.open("+.png").resize((100,80))
pic8 = ImageTk.PhotoImage(image8)
b8=Button(window,image=pic8,command=lambda button="+":cal(button)).place(x=450,y=320)

#third row

image9 = Image.open("4.png").resize((100,80))
pic9 = ImageTk.PhotoImage(image9)
b5=Button(window,image=pic9,command=lambda button="4":cal(button)).place(x=0,y=440)

image10 = Image.open("5.png").resize((100,80))
pic10 = ImageTk.PhotoImage(image10)
b10=Button(window,image=pic10,command=lambda button="5":cal(button)).place(x=150,y=440)

image11 = Image.open("6.png").resize((100,80))
pic11 = ImageTk.PhotoImage(image11)
b11=Button(window,image=pic11,command=lambda button="6":cal(button)).place(x=300,y=440)

image12 = Image.open("=.png").resize((100,80))
pic12 = ImageTk.PhotoImage(image12)
b12=Button(window,image=pic12,command=lambda button="=":answer()).place(x=450,y=440)

#fourth row

image13 = Image.open("1.png").resize((100,80))
pic13 = ImageTk.PhotoImage(image13)
b13=Button(window,image=pic13,command=lambda button="1":cal(button)).place(x=0,y=550)

image14 = Image.open("2.png").resize((100,80))
pic14 = ImageTk.PhotoImage(image14)
b14=Button(window,image=pic14,command=lambda button="2":cal(button)).place(x=150,y=550)

image15 = Image.open("3.png").resize((100,80))
pic15 = ImageTk.PhotoImage(image15)
b15=Button(window,image=pic15,command=lambda button="3":cal(button)).place(x=300,y=550)

image16 = Image.open("0.png").resize((100,80))
pic16 = ImageTk.PhotoImage(image16)
b16=Button(window,image=pic16,command = lambda button="0":cal(button)).place(x=450,y=550)

"""kkkkkkkkkkkkkkkkkkkkkkkkkkkkkkkkkkkkkkkkkkkkkkkkkkkkkkkkkkkkkkkkkkkkkkkkkkkkkk
kkkkkkkkkkkkkkkkkkkkkkkkkkkkkkkkkkkkkkkkkkkkkkkkkkkkkkkkkkkkkkkkkkkkkkkkkkkkkkkkkk
kkkkkkkkkkkkkkkkkkkkkkkkkkkkkkkkkkkkkkkkkkkkkkkkkkkkkkkkkkkkkkkkkkkkkkkkkkkkkkkkkk"""
     

window.mainloop()
```
